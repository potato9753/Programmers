```python
def solution(dirs):
    answer=0
    a = 5
    b = 5
    i=0
    dic = {}
    while(i<len(dirs)):
        if dirs[i] == 'L':
            if b-1<0:
                i+=1
                continue
            else:
                if (a,b) in dic:
                    if (a,b-1) in dic:
                        if 'L' not in dic[a,b] and 'R' not in dic[a,b-1]:
                            answer+=1
                            dic[a,b].append('L')
                    else:
                        if 'L' not in dic[a,b]:
                            answer+=1
                            dic[a,b].append('L')
                else:
                    if (a,b-1) in dic:
                        if 'R' not in dic[a,b-1]:
                            answer+=1
                            dic[a,b]=['L']
                    else:
                        answer+=1
                        dic[a,b]=['L']
                b-=1
        elif dirs[i] =='R':
            if b+1>10:
                i+=1
                continue
            else:
                if (a,b) in dic:
                    if (a,b+1) in dic:
                        if 'R' not in dic[a,b] and 'L' not in dic[a,b+1]:
                            answer+=1
                            dic[a,b].append('R')
                    else:
                        if 'R' not in dic[a,b]:
                            answer+=1
                            dic[a,b].append('R')
                else:
                    if (a,b+1) in dic:
                        if 'L' not in dic[a,b+1]:
                            answer+=1
                            dic[a,b]=['R']
                    else:
                        answer+=1
                        dic[a,b]=['R']
                b+=1
        elif dirs[i] =='U':
            if a-1<0:
                i+=1
                continue
            else:
                if (a,b) in dic:
                    if (a-1,b) in dic:
                        if 'U' not in dic[a,b] and 'D' not in dic[a-1,b]:
                            answer+=1
                            dic[a,b].append('U')
                    else:
                        if 'U' not in dic[a,b]:
                            answer+=1
                            dic[a,b].append('U')              
                else:
                    if (a-1,b) in dic:
                        if 'D' not in dic[a-1,b]:
                            answer+=1
                            dic[a,b]=['U']
                    else:
                        answer+=1
                        dic[a,b]=['U']
                a-=1
        elif dirs[i] =='D':
            if a+1>10:
                i+=1
                continue
            else:
                if (a,b) in dic:
                    if (a+1,b) in dic: 
                        if 'D' not in dic[a,b] and 'U' not in dic[a+1,b]:
                            answer+=1
                            dic[a,b].append('D')
                    else:
                        if 'D' not in dic[a,b]:
                            answer+=1
                            dic[a,b].append('D')               
                else:
                    if (a+1,b) in dic: 
                        if 'U' not in dic[a+1,b]:
                            answer+=1
                            dic[a,b]=['D']
                    else:
                        answer+=1
                        dic[a,b]=['D']

                a+=1
        i+=1
    return answer
```
