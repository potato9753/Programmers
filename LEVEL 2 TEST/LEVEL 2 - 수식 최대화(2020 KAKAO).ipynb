```python
def cut(arr):
    a = ''
    ex_=[]
    for i in arr:
        if i =='-' or i=='+' or i=='*':
            ex_.append(a)
            ex_.append(i)
            a=''
        else:
            a+=i
    if len(a)>0:
        ex_.append(a)
    return ex_

def minus(arr):
    i = 0
    a=[]
    while i<len(arr):
        if arr[i] == '-':
            a.append(str(int(a.pop())-int(arr[i+1])))
            i+=1
        else:
            a.append(arr[i])
        i+=1
    return a

def multi(arr):
    i = 0
    a=[]
    while i<len(arr):
        if arr[i] == '*':
            a.append(str(int(a.pop())*int(arr[i+1])))
            i+=1
        else:
            a.append(arr[i])
        i+=1
    return a

def plus(arr):
    i = 0
    a=[]
    while i<len(arr):
        if arr[i] == '+':
            a.append(str(int(a.pop())+int(arr[i+1])))
            i+=1
        else:
            a.append(arr[i])
        i+=1
    return a
def solution(expression):
    ex = cut(expression)
    num = []
    answer = 0
    for i in range(6):
        if i==0:
            a = ex
            a = multi(a)
            a = plus(a)
            a = minus(a)
            num.append(abs(int(a[0])))
        elif i==1:
            a = ex
            a = multi(a)
            a = minus(a)
            a = plus(a)
            num.append(abs(int(a[0])))
        elif i==2:
            a = ex
            a = minus(a)
            a = multi(a)
            a = plus(a)
            num.append(abs(int(a[0])))
        elif i==3:
            a = ex
            a = minus(a)
            a = plus(a)
            a = multi(a)
            num.append(abs(int(a[0])))
        elif i==4:
            a = ex
            a = plus(a)
            a = multi(a)
            a = minus(a)
            num.append(abs(int(a[0])))
        elif i==5:
            a = ex
            a = plus(a)
            a = minus(a)
            a = multi(a)
            num.append(abs(int(a[0])))
    answer = max(num)
    return answer
```
